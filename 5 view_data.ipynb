```python
import pandas as pd
df = pd.read_csv("arxiv.tsv", sep="\t", encoding="utf-8")
```

```python
df
```

```python
import os
from urllib.parse import urlparse

df['Week'] = pd.to_datetime(df['Week'])
os.makedirs('arxiv', exist_ok=True)

def make_hyperlink(url):
    if pd.isna(url) or not isinstance(url, str) or url.strip() == "":
        return ""
    url = url.replace('/abs/', '/pdf/')
    domain = urlparse(url).netloc
    if not domain:
        domain = url
    display = domain.replace('www.', '').replace('.com', '').replace('.org', '')
    return f"[{display}]({url})"

def make_twitter_link(url):
    if pd.isna(url) or not isinstance(url, str) or url.strip() == "":
        return ""
    return f"[Twitter]({url})"

for (year, quarter), group in df.groupby([df['Week'].dt.year, df['Week'].dt.quarter]):
    group = group.sort_values(by='Week', ascending=False)
    group['Week'] = group['Week'].dt.date
    group['Paper-Link'] = group['Paper-Link'].apply(make_hyperlink)
    group['Tweet-Link'] = group['Tweet-Link'].apply(make_twitter_link)
    group[['Week', 'title', 'category', 'abstract', 'Paper-Link', 'Tweet-Link']].to_markdown(
        f'arxiv/arxiv_{year}_Q{quarter}.md', index=False
    )
```
