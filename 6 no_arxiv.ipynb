```python
import pandas as pd
df = pd.read_csv("papers_clean.tsv", sep="\t", encoding="utf-8")
```

```python
df = df[df["has_arxiv"] == False].reset_index(drop=True)
df
```

```python
import os
df['Week'] = pd.to_datetime(df['Week'])
os.makedirs('no_arxiv', exist_ok=True)
for (year, quarter), group in df.groupby([df['Week'].dt.year, df['Week'].dt.quarter]):
    group = group.sort_values(by='Week', ascending=False)
    group['Week'] = group['Week'].dt.date
    group[['Week', 'Description', 'Paper-Link', 'Tweet-Link', 'Other-Links']].to_markdown(f'no_arxiv/{year}_Q{quarter}.md', index=False)
```
