```python
import requests
from IPython.core.display import display, HTML

url = 'https://arxiv.org/abs/2506.17298'
response = requests.get(url)
if response.status_code == 200:
    print("Successfully fetched the page.")
    
with open("arxiv_page.html", "w", encoding="utf-8") as file:
    file.write(response.text)
```

```python
import re

input_file = "arxiv_page.html"

with open(input_file, "r", encoding="utf-8") as f:
    html = f.read()

pattern = re.compile(
    r'(<div class="leftcolumn">.*?<div style="clear:both;"></div>)',
    re.DOTALL
)

match = pattern.search(html)
if match:
    extracted = match.group(1)

print(extracted)
```

```python
from arxiv import arxiv_to_json

arxiv_id, data = arxiv_to_json(extracted)
```
